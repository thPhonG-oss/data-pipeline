```python
from vnstock_data import Finance

# Khởi tạo đối tượng
fin_nonfinancial = Finance(symbol='HPG', period='quarter', source='vci')
fin_bank = Finance(symbol='MBB', period='quarter', source='vci')
fin_stock = Finance(symbol='VND', period='quarter', source='vci')
fin_insurrance = Finance(symbol='PVI', period='quarter', source='vci')
```

```python
fin_nonfinancial.balance_sheet(lang='vi', style='readable', get_all=True).to_csv('balance_sheet_nonfinancial.csv', index=False, encoding='utf-8-sig')
fin_bank.balance_sheet(lang='vi', style='readable', get_all=True).to_csv('balance_sheet_bank.csv', index=False, encoding='utf-8-sig')
fin_stock.balance_sheet(lang='vi', style='readable', get_all=True).to_csv('balance_sheet_stock.csv', index=False, encoding='utf-8-sig')
fin_insurrance.balance_sheet(lang='vi', style='readable', get_all=True).to_csv('balance_sheet_insurrance.csv', index=False, encoding='utf-8-sig')
```
