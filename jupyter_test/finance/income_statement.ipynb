```python
from vnstock_data import Finance

# Khởi tạo đối tượng
fin_nonfinancial = Finance(symbol='HPG', period='quarter', source='vci')
fin_bank = Finance(symbol='MBB', period='quarter', source='vci')
fin_stock = Finance(symbol='VND', period='quarter', source='vci')
fin_insurrance = Finance(symbol='PVI', period='quarter', source='vci')
```

```python
fin_nonfinancial.income_statement(lang='vi', style='readable', get_all=True).to_csv('income_statement_nonfinancial.csv', index=False, encoding='utf-8-sig')
fin_bank.income_statement(lang='vi', style='readable', get_all=True).to_csv('income_statement_bank.csv', index=False, encoding='utf-8-sig')
fin_stock.income_statement(lang='vi', style='readable', get_all=True).to_csv('income_statement_stock.csv', index=False, encoding='utf-8-sig')
fin_insurrance.income_statement(lang='vi', style='readable', get_all=True).to_csv('income_statement_insurrance.csv', index=False, encoding='utf-8-sig')
```
