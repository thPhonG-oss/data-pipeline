```python
from vnstock_data import Company
company = Company(symbol='VNZ', source='vci')

company.overview().to_csv('overview.csv', index=False, encoding='utf-8-sig')
```

```python
# ban lãnh đạo, filter_by= [working, resig, all]
company.officers(filter_by='working').to_csv('officers.csv', index=False, encoding='utf-8-sig')
```

```python
# Danh sách công ty con, công ty liên kết
# tham số filter_by
# bằng 'all' (mặc định), lấy cả công ty con, công ty liên kết
# bằng 'subsidiary', chỉ lấy thông tin công ty con (sở hữu từ 50% trở lên), không bao gồm công ty liên kết
company.subsidiaries(filter_by='subsidiary').to_csv('subsidiaries.csv', index=False, encoding='utf-8-sig')
company.subsidiaries(filter_by='all').to_csv('all.csv', index=False, encoding='utf-8-sig')
```

```python
# Sự kiện công ty
company.events().to_csv('events.csv', index=False, encoding='utf-8-sig')
```

```python
# Tin tức công ty
company.news().to_csv('news.csv', index=False, encoding='utf-8-sig')
```

```python
# Báo cáo phân tích liên quan đến công ty
company.reports().to_csv('reports.csv', index=False, encoding='utf-8-sig')
```

```python
# Tình hình tài chính
company.ratio_summary().to_csv('ratio_summary.csv', index=False, encoding='utf-8-sig')
```

```python
# Thống kê giao dịch
company.trading_stats().to_csv('trading_tats.csv', index=False, encoding='utf-8-sig')
```
